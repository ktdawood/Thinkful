# Two Continuous Variables

```python
# Plot 1
sns.lmplot("FLIGHT_COUNT", "AVG_INTERVAL", data = df, fit_reg = False, palette = 'pastel', size = 5)
```

```python
# Plot 2
sns.jointplot(x="FLIGHT_COUNT", y="AVG_INTERVAL", data=df, kind="reg", size = 8)
```

# Continuous Variable and Categorical Variable

```python
# Plot 1

sns.set(style = 'darkgrid')
g = sns.factorplot(x="GENDER", y="AGE", data=df[lambda df: df['AGE'] < 500],
                   size=6, kind="bar", palette="pastel",ci=95)
plt.show()
```

```python
# Plot2

sns.stripplot(x="GENDER", y="AGE", data=df[lambda df: df['AGE'] < 500], jitter=True, palette = 'pastel')
```

```python
#Plot 3

sns.boxplot(x="GENDER", y="AGE", data=df[lambda df: df['AGE'] < 500], palette = 'pastel')
```

```python
# Plot 4
sns.pointplot(x="GENDER", y="AGE", data=df[lambda df: df['AGE'] < 500], palette = 'pastel', ci = 99.99)
```

```python
# Plot 5
sns.set(style = 'darkgrid')
g = sns.factorplot(x="GENDER", y="AGE", data=df[lambda df: df['AGE'] < 500],
                   size=6, kind="violin", palette="pastel")
plt.show()
```

```python
# Plot 6:
sns.set(style = 'darkgrid')
g = sns.factorplot(x="GENDER", y="AGE", data=df[lambda df: df['AGE'] < 500],
                   size=6, kind="boxen", palette="pastel")
plt.show()
```

# Illustration on Plots:

1. distplot: Plots a univariate distribution of observations and an approximate density of the dataset

2. boxplot: Shows the distribution of quantitative data which facilitates comparisons between variables.

3. lmplot: Seaborn lmplot is a 2D scatter plot with a regression line. 

4. jointplot: Plots two variables with bivariate and univariate graphs.

5. stripplot: Shows all observations and a representation of some distribution

6. violinplot: A violin plot displays both density and probability distribution of a numerical variable.
    Pros:
    Shows summary statistics of a given variable.
    Unlike a box plot, shows the full distribution of the data. This allows for easy spotting of multimodal data.
    Cons:
    Violin plots are less known to wide audience meaning it may be harder to grasp.
